Import needed Library and Spliting sata into 2 part male and female

```python
from matplotlib import pyplot as plt
import pandas as pd

dataset = pd.read_csv("weight-height.csv")

male = dataset[dataset['Gender'] == 'Male']
female = dataset[dataset['Gender'] == 'Female']
```

count all neighbours in range k to see if who have more member in range **k** male or female

```python
userData = [60, 170]
k = 30

import math

distance = lambda x1, y1, x2, y2 : math.sqrt(math.pow(x1 - x2, 2) + math.pow(y1 - y2, 2))
 
maleScore = 0
for index in range(len(male)):
    magnitude = distance(male["Height"][index], male["Weight"][index], userData[0], userData[1])
    if magnitude <= k:
        maleScore += 1

femaleScore = 0
for index in range(len(male), len(female) + len(male)):
    magnitude = distance(female["Height"][index], female["Weight"][index], userData[0], userData[1])
    if magnitude <= k:
        femaleScore += 1
```

```python
fig, ax = plt.subplots()

ax.plot(male["Height"], male["Weight"], 'o', color="blue")
ax.plot(female["Height"], female["Weight"], 'D', color="red")
ax.set(xlabel='Height', ylabel='Weight')
ax.grid()

ax.plot(userData[0], userData[1], 'v', color="green")
print("Male" if maleScore > femaleScore else "Female" if maleScore - Femalescore != 0 else "Non-binary")
```
